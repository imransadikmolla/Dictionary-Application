```python

from tkinter import *
from PyDictionary import PyDictionary

#Create Objects

dictionary = PyDictionary()

rootword = Tk()

#Set geometry.

rootword.geometry("500x500")

#Set title

rootword.title("Dictionary created by Imran Sadik Molla")

#Set resize

#rootword.resizable(0,0)

# Function calling

def dict():

   meaning.config(text=dictionary.meaning(word.get())['Noun'][0])
   synonym.config(text=dictionary.synonym(word.get()))
   antonym.config(text=dictionary.antonym(word.get()))

# Labels, Button and Frames 
Label(rootword, text="Dictionary", font=("Helvetica 20 bold"), fg="Green").pack(pady=10)
# Frane 

frame =Frame (rootword)

Label(frame, text="Type Word", font=("Helvetica 15 bold")).pack(side=LEFT)
word = Entry(frame, font=("Helvetica 15 bold"))
word.pack()
frame.pack(pady=15)



# Frame 1
frame1= Frame(rootword)
Label(frame1, text="Meaning :- ", font=("Helvetica 10 bold")).pack(side=LEFT)
meaning = Label(frame1, text="", font=("Helvetica 10"))
meaning.pack()
frame1.pack(pady=15)



# Frame 2
frame2= Frame(rootword)
Label(frame2, text="Synonym :- ", font=("Helvetica 10 bold")).pack(side=LEFT)
synonym=Label(frame2, text="", font=("Helvetica 10"))
synonym.pack()
frame2.pack(pady=15)


# Frane 3
frame3= Frame(rootword)
Label(frame3, text="Antonym :- ", font=("Helvetica 10 bold")).pack(side=LEFT)
antonym = Label(frame3, text="", font=("Helvetica 10"))
antonym.pack(side=LEFT)
frame3.pack(pady=15) 


Button(rootword, text="Submit", font=("Helvetica 15 bold"), command=dict).pack()
rootword.mainloop()
```
